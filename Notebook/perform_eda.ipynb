```python
got_para = ['Longitude', 'Latitude', 'Temp_2m', 'MaxTemp_2m', 'MinTemp_2m', 'Pressure', 'Humidity_2m', 'WindSpeed_10m']
use_df = df[got_para]
use_df.sample(1)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(17,5))
plt.subplot(121)
sns.histplot(df['Temp_2m'], bins=30, kde=True, color='green')
plt.subplot(122)
sns.histplot(data=use_df, x='Temp_2m', kde=True)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x=df['Humidity_2m'], y=df['Pressure'])
plt.title('Pressure Vs Humidity')
plt.show()
```

```python
sns.histplot(df[['Temp_2m', 'MaxTemp_2m', 'MinTemp_2m']], kde=True, bins=30)
plt.xlabel('Temp(degree celcius)')
plt.show()
```

```python
district_temp = df.groupby('District')[['MaxTemp_2m']].max()

district_temp = district_temp.reset_index()
district_temp
district_humidity = df.groupby('District')['Humidity_2m'].max()
district_humidity = district_humidity.reset_index()
district_lang = df.groupby('District')[['Longitude','Latitude']].mean()
district_temp_map = pd.merge(district_lang,district_temp,on='District')
district_temp_map
district_map = pd.merge(district_temp_map,district_humidity,on="District")
district_map
```

```python
import folium

# Create Map Centered at a Location
m = folium.Map(location=[27.7172, 85.3240], zoom_start=5)

# Add Markers
for index, row in district_map.iterrows():
    folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=f"district name {row['District']}<br> temperature{row['MaxTemp_2m']}<br> high humidity{row['Humidity_2m']}"
    ).add_to(m)


# Show Map
#m
```

```python
data = df
```

```python
plt.figure(figsize=(14,6))
plt.plot(data['Precip'], label='Precipitation', color='blue')
plt.plot(data['Temp_2m'], label='Temperature', color='orange')
plt.title('Precipitation and Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
sns.scatterplot(x='Latitude', y='Longitude', data=data, hue='Precip', palette='viridis', size='Temp_2m', sizes=(20,200))
plt.title('Distrubution of precipitation and temperature')
plt.show()
```

```python
corr_data = data.drop('District', axis=1)
correction_matrix = corr_data.corr()

plt.figure(figsize=(22,15))
sns.heatmap(correction_matrix, annot=True, cmap='coolwarm', linewidth=.1)
plt.show()
```
